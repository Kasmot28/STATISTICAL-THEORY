2.) Analyze the data to provide the hypothesis testing conclusion. What is the p-value for your test? What is your recommendation for the researcher?

```python
import numpy as np
import scipy.stats as stats

Males = [12, 4, 11, 13, 11, 7, 9, 10, 10, 7, 7, 12, 6, 9, 15, 10, 11, 12, 7, 8, 8, 9, 11, 10, 9, 10, 9, 9, 7, 9, 11, 7, 10, 10, 11, 9, 12, 12, 8, 13, 9, 10, 8, 11, 10, 13, 13, 9, 10, 13]
Females = [11, 9, 7, 10, 9, 10, 10, 7, 9, 10, 11, 8, 9, 6, 11, 10, 7, 9, 12, 14, 11, 12, 12, 8, 12, 12, 9, 10, 11, 7, 12, 7, 9, 8, 11, 10, 8, 13, 8, 10, 9, 9, 9, 11, 9, 9, 8, 9, 12, 11]

t_stat, p_value = stats.ttest_ind(Males, Females)
print("T-test:", t_stat)
print("P-value:", p_value)

if p_value < 0.05: 
    print("we are rejecting null hypothesis")
else:
    print("we are accepting null hypothesis")
```

Since in the result, p-value resulted greater than 0.05, therefore the researcher fail to reject the null hypothesis, with this result, we can say that 
there is no difference in the mean time spent on cellphones between Males and Females in the sample.

4.) (10 points). What is the 95% confidence interval for the population mean of each gender category, and what is the 95% confidence interval for the difference between the means of the two populations?

```python
def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    stderr = stats.sem(data)
    interval = stderr * stats.t.ppf((1 + confidence) / 2., n-1)
    return mean - interval, mean + interval

# 95% CI for males
ci_males = confidence_interval(males)
print("95% Confidence Interval for Males:", ci_males)

# 95% CI for females
ci_females = confidence_interval(females)
print("95% Confidence Interval for Females:", ci_females)

# CI for the difference in means
mean_diff = np.mean(males) - np.mean(females)
se_diff = np.sqrt(np.var(males, ddof=1)/len(males) + np.var(females, ddof=1)/len(females))
ci_diff = (mean_diff - 1.96 * se_diff, mean_diff + 1.96 * se_diff)

print("95% Confidence Interval for Difference in Means:", ci_diff)
```
