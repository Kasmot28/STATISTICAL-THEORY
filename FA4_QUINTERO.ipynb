```python
import numpy as np

female_heights = {
    "Normal": [67, 69, 70, 62, 63, 67, 65, 59, 68, 66, 60, 65, 67, 63, 64, 65, 69, 60, 61, 67, 66, 64, 65, 68, 66, 62, 64, 67, 70, 63, 66, 65, 63, 66, 61, 65, 63, 63, 66, 66, 64, 64, 62, 67, 62, 66, 63, 66, 65, 63]
}

def calculate_moments(data, value):
    centered_values = np.array(data) - value
    return {
        "1st Moment": np.mean(centered_values),
        "2nd Moment": np.mean(centered_values**2),
        "3rd Moment": np.mean(centered_values**3),
        "4th Moment": np.mean(centered_values**4)
    }

heights = female_heights["Normal"]
moments = calculate_moments(heights, 75)

print("Moments about 75:")
for moment, value in moments.items():
    print(f"  {moment}: {value:.2f}")

```
